```python
from urllib.request import urlopen
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd
import re
import numpy as np
import time

def getAndParseURL(url):
    result=requests.get(url)
    soup=bts(result.text,"html.parser")
    return soup

pages=[]
#pages=["https://www.arabam.com/ikinci-el/otomobil/volkswagen-golf?sort=startedAt.desc&take=50"]
for page in range(11,21):
    pages.append("https://www.arabam.com/ikinci-el/otomobil/volkswagen-golf?sort=startedAt.desc&take=50&page=" +str (page))

pages
```

```python
cars=[]

for page in pages:
    html=getAndParseURL(page)
    for carlink in html.findAll("td",{"class":"listing-modelname pr"}):
        cars.append("https://www.arabam.com/"+carlink.a.get("href"))

cars
```

```python
len(cars)
```

```python
features  = []
for carl in cars:
    html = getAndParseURL(carl)

    try:
        price = html.find("div", {"class":"product-price"}).text.replace('TL', '').replace('.', '').strip()
    except:
        price = np.nan
    try:
        brand = html.find("div", {"class":"product-properties"}).find(string=re.compile("Marka")).findNext().text.strip()
    except:
        brand = np.nan
    try:
        series = html.find("div", {"class":"product-properties"}).find(string=re.compile("Seri")).findNext().text.strip()
    except:
        series = np.nan
    try:
        model = html.find("div", {"class":"product-properties"}).find(string=re.compile("Model")).findNext().text.strip()
    except:
        model = np.nan
    try:
        year = html.find("div", {"class":"product-properties"}).find(string=re.compile("Yl")).findNext().text.strip()
    except:
        year = np.nan
    try:
        km = html.find("div", {"class":"product-properties"}).find(string=re.compile("Kilometre")).findNext().text.replace('km', '').replace('.', '').strip()
    except:
        km = np.nan
    try:
        gear_type = html.find("div", {"class":"product-properties"}).find(string=re.compile("Vites Tipi")).findNext().text.strip()
    except:
        gear_type = np.nan
    try:
        fuel = html.find("div", {"class":"product-properties"}).find(string=re.compile("Yakt Tipi")).findNext("div").text.strip()
    except:
        fuel = np.nan
    try:
        body_type = html.find("div", {"class":"product-properties"}).find(string=re.compile("Kasa Tipi")).findNext().text.strip()
    except:
        body_type = np.nan

    try:
        color = html.find("div", {"class":"product-properties"}).find(string=re.compile("Renk")).findNext().text.strip()
    except:
        color = np.nan

    try:
        from_ = html.find("div", {"class":"product-properties"}).find(string=re.compile("Kimden")).findNext().text.strip()
    except:
        from_ = np.nan
    try:
        wheel_drive = html.find("div", {"class":"product-properties"}).find(string=re.compile("eki")).findNext().text.strip()
    except:
        wheel_drive = np.nan
    try:
        cylinder_number = html.find("div", {"class":"tab-content-car-information"}).find(string=re.compile("Silindir Says")).findNext().text.strip()
    except:
        cylinder_number = np.nan
    try:
        torque = html.find("div", {"class":"tab-content-car-information"}).find(string=re.compile("Tork")).findNext().text.replace('nm', '').strip()
    except:
        torque = np.nan
    try:
        eng_size = html.find("div", {"class":"product-properties"}).find(string=re.compile("Motor Hacmi")).findNext().text.replace('cc', '').strip()
    except:
        eng_size = np.nan
    try:
        engine_power = html.find("div", {"class":"product-properties"}).find(string=re.compile("Motor Gc")).findNext().text.replace('hp', '').strip()
    except:
        engine_power = np.nan
    try:
        max_power = html.find("div", {"class":"tab-content-car-information"}).find(string=re.compile("Maksimum G")).findNext().text.replace('rpm', '').strip()
    except:
        max_power = np.nan
    try:
        min_power = html.find("div", {"class":"tab-content-car-information"}).find(string=re.compile("Minimum G")).findNext().text.replace('rpm', '').strip()
    except:
        min_power = np.nan
    try:
        acceleration = html.find("div", {"class":"tab-content-car-information"}).find(string=re.compile("Hzlanma")).findNext().text.replace('sn', '').strip()
    except:
        acceleration = np.nan
    try:
        max_speed = html.find("div", {"class":"tab-content-car-information"}).find(string=re.compile("Maksimum Hz")).findNext().text.replace('km/s', '').strip()
    except:
        max_speed = np.nan
    try:
        average_fuel_consumption = html.find("div", {"class":"product-properties"}).find(string=re.compile("Ort. Yakt Tketimi")).findNext().text.replace('lt', '').strip()
    except:
        average_fuel_consumption = np.nan
    try:
        painted_changed = html.find("div", {"class":"product-properties"}).find(string=re.compile("Boya-deien")).findNext().text.strip()
        if painted_changed == "Belirtilmemi":
            painted_changed = np.nan
        if painted_changed == "Tamam orjinal":
            painted_changed = "0 deien, 0 boyal"

    except:
        painted_changed = np.nan

    features.append([price, brand, series, model, year, km, gear_type, fuel, body_type,color, from_, wheel_drive, cylinder_number, torque, eng_size, engine_power, max_power, min_power, acceleration, max_speed, average_fuel_consumption, painted_changed])


```

```python
columns = ['price_try', 'brand', 'series', 'model', 'year', 'km', 'gear_type', 'fuel', 'body_type','color' ,'from_', 'wheel_drive', 'cylinder_number', 'torque_nm', 'engSize_cc', 'engPower_hp', 'max_power_rpm', 'min_power_rpm', 'acceleration_100_sec', 'max_speed_kmh', 'avg_fuel_lt', 'painted_changed']
features
```

```python
df = pd.DataFrame.from_records(features, columns=columns)
df
```

```python
# price-discount varsa
# Verinin sa tarafndaki boluklar, yeni satr karakterlerini ve dzensizlikleri temizleyerek sadece sondaki saylar al

df['price_try'] = df['price_try'].str.strip().str.split().str[-1]
```

```python
#df.drop("Unnamed: 0",axis=1,inplace=True)
df["brand"].unique()
df.info()
```

```python
# Boya Deisen kolonlarn ayralm
df['painted'] = df['painted_changed'].str.extract(r'(\d+)\sboyal', expand=False)
df['changed'] = df['painted_changed'].str.extract(r'(\d+)\sdeien', expand=False)

# NaN deerlerini 0'a dntrme
df['painted'] = df['painted'].fillna(0).astype(int)
df['changed'] = df['changed'].fillna(0).astype(int)

# NaN deerlerini numpy.nan ile deitirme
#df['painted'] = df['painted'].replace('', np.nan)
#df['changed'] = df['changed'].replace('', np.nan)

df.drop(columns=['painted_changed'], inplace=True)
df
```
